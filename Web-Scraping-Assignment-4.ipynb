```python
# clicking on explore 
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/summary").click()
```

```python
# clicking on trending option
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a").click()
```

```python
title=[] # empty list
# scraping repository title 
for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']"):
    title.append(i.text)
title
```

```python
desc=[] # empty list
# scraping repository description
try:
    for i in driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']"):
        desc.append(i.text)
except NoSuchElementException:
    desc.append("--")
```

```python
# inserting '--' to selected index positions 
desc.insert(8,'--')
desc.insert(19,'--')
desc.insert(20,'--')
```

```python
desc
```

```python
# making empty lists    
repo_urls=[]
Contributors_count=[]
Language_used=[]
```

```python
#scraping repositories urls
repos=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in repos:
    repo_urls.append(i.get_attribute('href'))
time.sleep(2) 
```

```python
for i in repo_urls: # running loop
    driver.get(i)
    l=[]
    time.sleep(3)
    
    #scraping contributors count
    try:
        count=driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        Contributors_count.append(count.text)
    except NoSuchElementException:#handling no such element exception
        Contributors_count.append('No details available')
    time.sleep(2)
    
     #scraping languages used   
    languages=driver.find_elements_by_xpath("//li[@class='d-inline']//a//span[1]")
    if languages:    
        for i in languages:
            l.append(i.text)
    else:
        l.append('No languages used')
    Language_used.append(l)   
```

```python
github=pd.DataFrame({})            # making dataframe
github['Repository title']=title
github['Repository description']=desc
github['Contributors count']=Contributors_count
github['Languages used']=Language_used

github
```

```python
driver.close() # to close current browser window
```

```python
# Maximize current window
driver.maximize_window()
```

```python
# empty list
name=[]
year_span=[]
genre=[]
run_time=[]
ratings=[]
vote=[]
votes = [] 
```

```python
# scraping name
try:
    for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a"):
        name.append(i.text)
except NoSuchElementException:
    name.append("---")

# scraping year
try:
    for i in driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]"):
        year_span.append(i.text)
except NoSuchElementException:
    year_span.append("---")

# scraping genre
try:
    for i in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[5]"):
        genre.append(i.text)
except NoSuchElementException:
    genre.append("---")

# scraping run time
try:
    for i in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[3]"):
        run_time.append(i.text)
except NoSuchElementException:
    run_time.append("---")

# scraping ratings
try:
    for i in driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]"):
        ratings.append(i.text)
except NoSuchElementException:
    ratings.append("---")

# scraping votes
try:
    for i in driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[2]"):
        vote.append(i.text)
except NoSuchElementException:
    vote.append("---")
    
for i in range(1,len(vote),2):
    votes.append(vote[i].strip())
```
